4. Как найти самое часто встречаемое значение между двумя массивами? 

````python
import numpy as np

array1 = np.array([1, 2, 3, 4, 5, 1, 2])
array2 = np.array([2, 3, 4, 4, 4, 6, 1])

combined_array = np.concatenate((array1, array2))

values, counts = np.unique(combined_array, return_counts=True)

most_common_index = np.argmax(counts)

most_common_value = values[most_common_index]
frequency = counts[most_common_index]

print(f"Самое часто встречаемое значение: {most_common_value}, частота: {frequency}")
````

7. Создайте массив случайных чисел размером 10 и отсортируйте его

````python
import numpy as np

def generate_and_sort_random_array(size=10, low=0, high=100):
    random_array = np.random.randint(low, high, size)
    sorted_array = np.sort(random_array)
    
    return sorted_array

sorted_random_array = generate_and_sort_random_array(size=10, low=1, high=50)

print("Отсортированный массив:", sorted_random_array)

````

8. Создайте случайную матрицу 10х2 которая представляет собой декартовые координаты и преобразуйте их в полярные

````python
import numpy as np

def cartesian_to_polar(cartesian_coords):
    x = cartesian_coords[:, 0]
    y = cartesian_coords[:, 1]
    
    r = np.sqrt(x**2 + y**2)
    
    theta = np.arctan2(y, x)
    return np.column_stack((r, theta))

cartesian_coords = np.random.rand(10, 2) * 100
polar_coords = cartesian_to_polar(cartesian_coords)

print("Декартовые координаты (x, y):\n", cartesian_coords)
print("\nПолярные координаты (r, θ):\n", polar_coords)
````

6. Как физическая активность влияет на успеваемость?

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

file_path = 'student_perfomance.csv'
df = pd.read_csv(file_path)

sns.set(style="whitegrid")

mean_scores = df.groupby('Physical_Activity')['Exam_Score'].mean().reset_index()

plt.figure(figsize=(8, 6))
sns.barplot(data=mean_scores, x='Physical_Activity', y='Exam_Score', edgecolor='black')

plt.title('Average Exam Scores by Physical Activity Level', fontsize=16)
plt.xlabel('Physical Activity Level', fontsize=12)
plt.ylabel('Average Exam Score', fontsize=12)

for index, row in mean_scores.iterrows():
    plt.text(index, row.Exam_Score + 1, f'{row.Exam_Score:.1f}', color='black', ha='center')

plt.show()

print("Физическая активность не имеет значительного влияния на результаты экзаменов")
````
